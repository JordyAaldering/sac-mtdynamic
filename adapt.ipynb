```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid', rc={'ytick.left' : True})
pal = sns.color_palette(palette='icefire')
primary = pal[1]
secondary = pal[5]
highlight = pal[0]
```

```python
path = 'results_omen/adapt_nbody.csv'
df = pd.read_csv(path, sep=' ', names=['power', 'runtime', 'energy'])
df['power'] /= 1e6
df['energy'] *= 1e3
df
```

```python
print('Total time:', df['runtime'].sum())
```

```python
ax = sns.lineplot(df, x=df.index, y='power')
ax.set_xlabel('Iteration')
ax.set_ylabel('Power limit (W)')
ax.set_ylim((0,25))
plt.show()
```

```python
ax = sns.lineplot(df, x=df.index, y='energy')
ax.set_xlabel('Iteration')
ax.set_ylabel('Energy (mJ)')
ax.set_ylim((0,4))
plt.show()
```

```python
ax = sns.lineplot(df, x=df.index, y='runtime')
ax.set_xlabel('Iteration')
ax.set_ylabel('Runtime (s)')
ax.set_ylim((0,0.8))
plt.show()
```
