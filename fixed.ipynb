```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from glob import glob
```

# Matmul laptop

```python
frames = []

for path in glob('laptop//matmul_fixed_*.csv'):
    print('Read:', path)
    df = pd.read_csv(path)

    uids = df['uid'].unique()
    funs = ['iota', 'transpose', 'matmul']
    rename = dict(zip(uids, funs))
    df.replace({'uid': rename}, inplace=True)

    df['size'] = int(path.split('_')[2])

    frames.append(df)

df = pd.concat(frames)
for (name, group) in df.groupby('uid'):
    print('\n=====', name, '=====')
    print(group.describe())
```

```python
sns.barplot(df[df['uid'] == 'matmul'], x='threads', y='runtime')
plt.show()
```

```python
sns.barplot(df[df['uid'] == 'matmul'], x='threads', y='energy')
plt.show()
```

```python
sns.barplot(df[df['uid'] == 'transpose'], x='threads', y='runtime')
plt.show()
```

```python
sns.barplot(df[df['uid'] == 'transpose'], x='threads', y='energy')
plt.show()
```
